```python
import pandas as pd
import numpy as np
import torch


## PREPARATION OF THE DATA
biometry = pd.read_csv('/data/kpusteln/Fetal-RL/data_preparation/org_data/biometry_org.csv')

femur_biometry = biometry[['ID', 'FL', 'femur_ps']]
abdomen_biometry = biometry[['ID', 'AC', 'abdomen_ps']]
head_biometry = biometry[['ID', 'HC', 'head_ps']]

femur_biometry['ID'] = femur_biometry['ID'].astype(str)
abdomen_biometry['ID'] = abdomen_biometry['ID'].astype(str)
head_biometry['ID'] = head_biometry['ID'].astype(str)

for i in range(len(femur_biometry)):
    idx = femur_biometry.loc[i, 'ID']
    new_idx = idx + '_3'
    femur_biometry.loc[i, 'ID'] = new_idx
    
for i in range(len(abdomen_biometry)):
    idx = abdomen_biometry.loc[i, 'ID']
    new_idx = idx + '_2'
    abdomen_biometry.loc[i, 'ID'] = new_idx
    
for i in range(len(head_biometry)):
    idx = head_biometry.loc[i, 'ID']
    new_idx = idx + '_1'
    head_biometry.loc[i, 'ID'] = new_idx
```

```python
head_biometry = head_biometry.rename(columns={'HC': 'measure', 'head_ps': 'ps', 'ID': 'video'})
abdomen_biometry = abdomen_biometry.rename(columns={'AC': 'measure', 'abdomen_ps': 'ps', 'ID': 'video'})
femur_biometry = femur_biometry.rename(columns={'FL': 'measure', 'femur_ps': 'ps', 'ID': 'video'})
```

```python
new_data = pd.concat([femur_biometry, abdomen_biometry, head_biometry], axis=0)
```

```python
data = pd.read_csv('/data/kpusteln/Fetal-RL/data_preparation/outputs/labels_corrected.csv')
merged = pd.merge(data, new_data, on='video')
```

```python
head = data['index'].str.split('_').str[1] == '1'
abdomen = data['index'].str.split('_').str[1] == '2'
femur = data['index'].str.split('_').str[1] == '3'
```

```python
data_abdomen = data[abdomen]
data_head = data[head]
data_femur = data[femur]
```

```python
# abdomen
ID = data_abdomen['index'].str.split('_').str[0]
ID = ID.astype(int)
data_abdomen['ID'] = ID
data_merged = pd.merge(data_abdomen, biometry_abdomen, on='ID')
data_merged.drop(columns = 'ID', inplace=True)
data_merged.to_csv('/data/kpusteln/Fetal-RL/data_preparation/data_biometry/abdomen_biometry.csv', index=False)
```

```python
# head
ID = data_head['index'].str.split('_').str[0]
ID = ID.astype(int)
data_head['ID'] = ID
data_merged = pd.merge(data_head, biometry_head, on='ID')
data_merged.drop(columns = 'ID', inplace=True)
data_merged.to_csv('/data/kpusteln/Fetal-RL/data_preparation/data_biometry/head_biometry.csv', index=False)
```

```python
# femur
ID = data_femur['index'].str.split('_').str[0]
ID = ID.astype(int)
data_femur['ID'] = ID
data_merged = pd.merge(data_femur, biometry_femur, on='ID')
data_merged.drop(columns = 'ID', inplace=True)
data_merged.to_csv('/data/kpusteln/Fetal-RL/data_preparation/data_biometry/femur_biometry.csv', index=False)
```
